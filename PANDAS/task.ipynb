```python
#task1.1
#Determine the top most 3 cuisines in the dataset
import numpy as np
import pandas as pd
df = pd.read_csv('Dataset .csv')
df
```

```python
#Determine the top most 3 cuisines in the dataset
df['Cuisines'] = df['Cuisines'].astype(str).str.strip()
top_3C =df['Cuisines'].value_counts().head(3)
print(top_3C)
```

```python
#task 1.2
#Calculate percentage of the restaurants that serve each of the top cuisines
total_restaurants = len(df)
cuisine_percentage = (df['Cuisines'].value_counts(normalize=True)*100).round(2)
print(f"Total Restaurants Analyzed : {total_restaurants}\n")
print("Percentage of the Restaurants serving each of the top cuisines are: ")
for Cuisines, percentage in cuisine_percentage.head(3).items():
    print(f" {Cuisines}: {percentage}% of all Restaurants")
```

```python
#identify the city with the highest number of restaurants in the dataset 
df['City'] = df['City'].astype(str).str.strip()
top_city = df['City'].max()
restaurant_count = df['City'].value_counts().max()
print(f"The city with the highest number of restaurants is: {top_city}")
print(f"Total number of restaurants in that city: {restaurant_count}")
```

```python
#task1.3
#to calculate average rating of the restaurants in each city
df['City'] = df['City'].astype(str).str.strip()
city_avg_rating = df.groupby('City')['Aggregate rating'].mean().round(2)
print(city_avg_rating.head(10))
```

```python
#determine the city with the highest average rating
top_rating = city_avg_rating.sort_values(ascending=False)
print(top_rating)
```

```python
#task 1.4
#Determine the percentage of the restaurants that offer online delivery
percentage = (df['Has Online delivery'] == 'Yes').mean() *100
print(f"{percentage: .2f}%")
```

```python
#compare the average rating of restaurants with and without online delivery
delivery_comparison = df.groupby('Has Online delivery')['Aggregate rating'].mean()

print(delivery_comparison)
```

```python
#task2.1
#analyze the distribution of aggregate ratings and determine the most common rating range 
import matplotlib.pyplot as plt

df['Aggregate rating'].describe()
cr = df['Aggregate rating'].quantile(0.75)
print(f"75% common range is :{cr: .2f}%")
mr = pd.cut(df['Aggregate rating'],bins=5).value_counts()
print(f"Most frequent range is: {mr}")
plt.figure(figsize=(10, 6))
df['Aggregate rating'].hist(bins=20, edgecolor='black', color='skyblue')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Rating Value')
plt.ylabel('Number of Restaurants')
plt.grid(False)
plt.show()
```

```python
#calculate the average no. of votes received by restaurants
avg_votes = df['Votes'].mean()
print(avg_votes)
```

```python
#task2.2
#identify most common combination of cuisines in the dataset
cmb = df['Cuisines'].value_counts().head(10)
print(cmb)
```

```python
#determine if certain cuisine combinations tend to have higher ratings
cuisine_stats = df.groupby('Cuisines')['Aggregate rating'].agg(['mean', 'count'])
top_performing_combos = cuisine_stats[cuisine_stats['count'] > 5].sort_values(by='mean', ascending=False)

print(top_performing_combos.head(10))
```

```python
#task2.3
#Plot the locations of restaurants on a map using longitude and latitude cordinates
df.plot(kind="scatter", x="Longitude", y="Latitude", alpha=0.4, figsize=(10,7), color='red')
plt.title("Spatial Distribution of Restaurants")
plt.show()
```

```python
#identify any pattern or clusters of restaurants in any specific area
maj_cluster = df.groupby('Locality').size()
maj_cluster = maj_cluster[maj_cluster > 50].sort_values(ascending=False)
print(maj_cluster)

rating_trend = df.groupby('Aggregate rating')['Votes'].mean()
plt.figure(figsize=(10, 5))
plt.plot(rating_trend.index, rating_trend.values, marker='o', linestyle='-', color='b')
plt.title('Pattern: Average Votes by Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Average Votes')
plt.grid(True)
plt.show()

df['Cost Bin'] = pd.qcut(df['Average Cost for two'], q=10, duplicates='drop').apply(lambda x: x.mid)
cost_trend = df.groupby('Cost Bin')['Aggregate rating'].mean()

plt.plot(cost_trend.index, cost_trend.values, marker='s', color='g')
plt.title('Link: Average Cost vs. Rating')
plt.xlabel('Average Cost for Two')
plt.ylabel('Average Rating')
plt.show()
```

```python
#task2.4
#to identify if there are any restaurant chains present in the dataset
restaurant_counts = df['Restaurant Name'].value_counts()
chains = restaurant_counts[restaurant_counts > 1]
print("Top 10 Restaurant Chains by number of outlets:")
print(chains.head(10))
```

```python
#Analyze the ratings and popularity of different restaurant chains
chain_analysis = df.groupby('Restaurant Name').agg({
    'Aggregate rating': 'mean',
    'Votes': 'sum',
    'Restaurant ID': 'count'
})

chain_analysis.columns = ['Avg_Rating', 'Total_Votes', 'Outlets']

chains = chain_analysis[chain_analysis['Outlets'] > 1].sort_values(by='Total_Votes', ascending=False)

print(chains.head(10))
```

```python
#task3.1
#Analyse the text reviews to identify the most positive and negative reviews

#using sentiment analysis
def quick_sentiment(text):
    text = str(text).lower()
    if any(word in text for word in ['good','great','amazing','delicious','excellent','very good']):
        return 'Positive'
    elif any(word in text for word in ['poor','bad','very poor','not good']):
        return 'Negative'
    return 'Neutral'

df['Basic_Sentiment'] = df['Rating text'].apply(quick_sentiment)
print("Sentiment Counts:")
print(df['Basic_Sentiment'].value_counts())
plt.figure(figsize=(7, 7))
df['Basic_Sentiment'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightgreen', 'lightgray', 'salmon'])
plt.title('Proportion of Sentiments')
plt.ylabel('') 
plt.show()
```

```python
#calculate the average length of reviews and explore if there is a relationship between reviews length and rating

df['Review_length'] = df['Rating text'].str.len()

avg_len = df['Review_length'].mean()
print(f"The average Review Length is: {avg_len: .2f} characters")


corr2 = df['Review_length'].corr(df['Aggregate rating'])
print(f"Correlation between length and rating: {corr2: .4f}")

plt.figure(figsize=(10, 6))
plt.scatter(df['Aggregate rating'], df['Review_length'], alpha=0.3, color='purple')
plt.title('Relationship: Review Length vs. Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Review Length (Characters)')
plt.show()
```

```python
#Task 3.2 Votes Analysis
#identify the restaurants with the highest And lowest no. of votes


high_votes = df.nlargest(10, 'Votes')[['Restaurant Name', 'Votes', 'Aggregate rating']]
print("Restaurants with Highest Votes:")
print(high_votes)


low_votes = df.nsmallest(10, 'Votes')[['Restaurant Name', 'Votes', 'Aggregate rating']]
print("\nRestaurants with Lowest Votes:")
print(low_votes)
```

```python
#Analyse if there is any correlation between the no. of votes and rating of restaurants
correlation = df['Votes'].corr(df['Aggregate rating'])

print(f"The correlation between Votes and Aggregate Rating is: {correlation:.4f}")

plt.figure(figsize=(10, 6))
plt.scatter(df['Votes'], df['Aggregate rating'], alpha=0.3, color='darkorange')

plt.title('Correlation: Number of Votes vs. Aggregate Rating')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#task 3.3
#Analyze if there is any relationship between the price range and availability of online delivery and table booking

delivery_pct = pd.crosstab(df['Price range'], df['Has Online delivery'], normalize='index') * 100
print("Online Delivery Availability by Price Range (%):")
print(delivery_pct)

booking_pct = pd.crosstab(df['Price range'], df['Has Table booking'], normalize='index') * 100
print("\nTable Booking Availability by Price Range (%):")
print(booking_pct)
```

```python
#determine fi higher priced restaurants are more likely to offer such services
service_analysis = df.groupby('Price range').agg({
    'Has Table booking': lambda x: (x == 'Yes').mean() * 100,
    'Has Online delivery': lambda x: (x == 'Yes').mean() * 100
})

service_analysis.columns = ['% Table Booking', '% Online Delivery']
print(service_analysis)
```
